#**DataFrame: Criação e Indexação**

```python
import pandas as pd
import numpy as np
```

##**Criando um DataFrame:**

```python
df = pd.DataFrame({"Peso": [72,80,60],"Altura": [180,170,165], "Idade": [26,19,15]})
display(df)
```

```python
#Outra Forma
dados = np.array([[72,180,26],[80,170,19],[60,165,15]])
df = pd.DataFrame(data = dados, columns = ['Peso', 'Altura', 'Idade'], index = [1,2,3]) #Ja mudando os indices
display(df)
```

##**Indexação:**

```python
#Acessando colunas
df['Peso'] #retorna uma Serie
```

```python
#Acessando Linhas
df.loc[1] #acessando a linha de indice 1
```

```python
#acessando elemento específico
df.loc[1, 'Peso']
```

```python
#outra forma
df['Altura'][1]
```

```python
#acessando duas ao mesmo tempo
display(df[['Peso', 'Altura']])
```

```python
#Usando o iloc => acessar os indices do dataframe usando apenas os numeros
#primeira linha x coluna
print(df.iloc[2,1]) #df.iloc[2][1]
```

##**Excluindo colunas**

```python
#para excluir uma coluna ou duas, usa-se o método drop
#O método drop retorna um novo dataframe sem a coluna ou linha excluida
'''df.drop(
  Parametros:
  axis = 0 => deleta linhas
  axis = 1 => deleta colunas
  sintaxe: drop(Número da linha, axis = 0)
                        ou
           drop("Nome da coluna", axis = 1)
)

'''
df2 = df2.drop('Idade', axis = 1)
df2.head()
#mas pode usar um parâmetro chamado inplace que é false como padrão
#Esse parametro pega o retorno e joga no dataFrame
```
