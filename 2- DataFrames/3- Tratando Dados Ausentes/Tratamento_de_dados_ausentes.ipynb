```python
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
dados = np.array([[1,      2, np.nan],
                  [4, np.nan, np.nan],
                  [7,       8,     9]])

dados_df = pd.DataFrame(dados, columns = "A B C".split())
```

```python
display(dados_df)
```

```python
#Usando o método isnull()
dados_df.isnull() #retorna uma máscara
#verifica quais dados estão auxentes
```

```python
dados_df.loc[1].isnull()
```

```python
dados_df['B'].isnull().sum() #soma a qauntidade de dadso auxentes da coluna B
```

```python
#Usando o método dropna
'''
parametros do dropna
axis = 0, elimina os valores nulos com relação a linha
axis = 1, elimina os valores nulos com relação a coluna
how = 'all', exclui todas as colunas ou linhas vazia
'''
dados_df2 = dados_df.dropna()
display(dados_df2)
```

```python
#Usando o fillna(), preenche os dados auxentes
dados_df['B'].fillna(0, inplace = True)
display(dados_df)
```

```python
dados_df.fillna(0, inplace = True)  #Vai pegar todo o dataframe
dados_df
```
