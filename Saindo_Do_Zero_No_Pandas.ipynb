#**Adicionar 1 coluna**

#**Adicionando uma linha**

#**Excluindo Linhas e Colunas**

```python
vendas_df2 = vendas_df2.drop("Imposto", axis = 1 ) # axis = 0 => eixo da linha; axis = 1 => eixo das colunas
display(vendas_df2)
```

```python
vendas_df2 = vendas_df2.drop(0 , axis = 0) #excluindo linhas, axis = 0
display(vendas_df2)
```

#**Valores Vazios**
- Deletar Linhas/Colunas vazias
- Deletar Linhas que possuem valores vazios
- Preencher valores vazios (média e último valor)


```python
#deletar linhas e colunas completamente vazias
vendas_df2.loc[:, 'Imposto'] = None  #Crieia uma colunas sem valores
vendas_df2 = vendas_df2.dropna( how = 'all', axis = 1 ) # vai excluir todas as colunas vazias ('Imposto')
display(vendas_df2)
```

```python
#deletar linhas que tem pelo menos um valor vazio
vendas_df2 = vendas_df2.dropna()
vendas_df2.shape #mostra que eu add uma colunas e foi excluida uma linha com valor vazio
```

```python
# preencher valorez vazios
# preenchendo com a média
n = vendas_df2['Comissão'].mean() #esse N pode assumir qualquer valor, o qual será colocado no lugar dos dados vazios
vendas_df2['Comissão'] = vendas_df2['Comissão'].fillna(n)
display(vendas_df2)
```

```python
display(vendas_df2.loc[vendas_df2['Código Venda'] == 7086])
```

```python
#Preencher com o último valor
vendas_df2 = vendas_df2.ffill()
```

##**Calcular Indicadores**
- Groupby
- Value Counts

```python
transacoes_loja = vendas_df2['ID Loja'].value_counts() #contar todos os valores
display(transacoes_loja)
```

```python
faturamento_produto = vendas_df2.groupby('Produto').sum() #digo qual coluna e depois a operação, nesse caso vou somar todas as colunas numéricas
display(faturamento_produto)
```

```python
faturamento_produto = vendas_df2[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```
