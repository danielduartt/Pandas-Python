```python
import numpy as np
import pandas as pd
```

#**Junção de Data Frames:**
- concat
- merge
- join

###Concat:

```python
dados = np.arange(1, 10).reshape(3, 3)
dados
```

```python
df1 = pd.DataFrame(dados, columns = 'A B C'.split())
df2 = pd.DataFrame(dados, columns = 'A B C'.split())
df3 = pd.DataFrame(dados, columns = 'D E F'.split())
```

```python
df1
```

```python
df2
```

```python
df3
```

#**Concat**
concatenar um abaixo do outro

```python
'''
parametros: pd.concat([lista dos dataframes que serão concatenados])
ignore_index = True/False (padrão: False) => concatenar os dataframes ajeitando os indices
axis = 1/0 (padrão: 0) => 0 para colocar na vertical, 1 para colocar na horizontal
'''
pd.concat([df1, df2]) #nesse caso coloca um abaixo do outro, contudo os indices continuam 0,1,2
```

```python
#concatenando e ajeitando os indices
pd.concat([df1, df2], ignore_index = True)
```

```python
#concatenando na horizontal
pd.concat([df1, df3], axis = 1 )
```

```python
pd.concat([df1,df2,df3], ignore_index = True)
```

###Merge (mesclando):

```python
#O merge vai misturar uma ou mais de uma das colunas
df1 = pd.DataFrame(dados, columns = 'A B C'.split())
df2 = pd.DataFrame(dados , columns ='D E C'.split())
df1
```

```python
df2
```

```python
#vou misturar os conteudos da coluna que é igual para os dois dataframes
#nesse caso é a coluna C
pd.merge(df1, df2, on = "C") #a C não vai ser duplicada
```

```python
df1 = pd.DataFrame(dados, columns = 'A B C'.split())
df2 = pd.DataFrame(dados , columns ='D B C'.split())
#agora vamos mesclar em Duas colunas igual
```

```python
pd.merge(df1, df2, on = ["B", "C"])
```

###Join (juntar)


```python
#join é um método dp obj dataframe
df1 = pd.DataFrame(dados, columns = 'A B C'.split(), index = [0, 1, 2])
df2 = pd.DataFrame(dados , columns ='D E F'.split(), index = [1, 2, 3])
'''
parametros:
how = (left, right ou outer), faz o join levando em consideração o indices da (esquerda,direita, completa com NaN os indices não presentes)
'''
df1.join(df2, how = 'outer')
```
