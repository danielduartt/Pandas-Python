```python
import numpy as np
import pandas as pd
```

#**Agrupamento de Dados**
- groupby

```python
df = pd.read_csv('/pesquisa-covid.csv')
display(df.head())
```

```python
df.shape
```

```python
df_covid = df.groupby('COVID-19') #agrupamento pela coluna 'COVID-19' calculando a média
df_covid.mean() #dados criados aleatóriamentes
```

```python
df_covid.sum()
```

#**Usando o DataSet do Titanic:**

```python
df_titanic = pd.read_csv('/titanic_train.csv')
display(df_titanic.head())
```

```python
df_survived = df_titanic.groupby("Survived")
```

```python
df_survived.mean()
```

```python
#percebe-se que as pesoas que sobreviveram estavam na primeira ou segunda classe
```

```python
df_sex = df_titanic.groupby("Sex").mean()
display(df_sex)
```

```python
df_sex_std = df_titanic.groupby("Sex").std()
df_sex_std
```
